```python
##Question 9: Python program to generate 1000 random numbers from a normal distribution
# answer : We generate samples from
"""𝑁(50,5) then compute the mean, standard deviation, and plot a histogram."""

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
samples = np.random.normal(loc=50, scale=5, size=1000)

sample_mean = np.mean(samples)
sample_std = np.std(samples, ddof=1)

print("Sample mean:", sample_mean)
print("Sample std:", sample_std)

plt.hist(samples, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of 1000 samples from N(50,5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
## Question 10: CLT application for daily sales & Python code
"""
We use CLT to estimate the population mean daily sales.
With 𝑛 = 20, the sample mean follows an approximately normal distribution.
A 95% CI is given by: 𝑥ˉ ± 𝑡0.975,𝑑𝑓=𝑛−1 ⋅ 𝑠/√𝑛
"""

import numpy as np
from scipy import stats

daily_sales = np.array([220,245,210,265,230,250,260,275,240,255,
                        235,260,245,250,225,270,265,255,250,260])

n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
sem = std_sales / np.sqrt(n)

t_crit = stats.t.ppf(0.975, df=n-1)
margin = t_crit * sem
ci_lower, ci_upper = mean_sales - margin, mean_sales + margin

print("Mean sales =", mean_sales)
print(f"95% CI = ({ci_lower}, {ci_upper})")
```
